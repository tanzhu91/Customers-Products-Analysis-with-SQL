```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df1 = pd.read_csv('../data/customers_high_credit_limit.csv')
df2 = pd.read_csv('../data/customers_low_credit_limit.csv')
```

```python
df3 = pd.read_csv('../data/low_in_stock_products.csv')
df4 = pd.read_csv('../data/high_in_stock_products.csv')
```

```python
df5 = pd.read_csv('../data/top_5_customers_by_revenue.csv')
df6 = pd.read_csv('../data/bottom_5_customers_by_revenue.csv')
```

```python
df7 = pd.read_csv('../data/top_sales.csv')
df8 = pd.read_csv('../data/lowest_sales.csv')
```

```python
#low in stock products
order = df3.sort_values(by='quantityInStock', ascending=True)['productName']
sns.barplot(data=df3,y='productName',x='quantityInStock',order=order)
plt.title('Barplot of Low in Stock Products')
plt.ylabel('Product Name')
plt.xlabel('Quantity in stock');
```

```python
#high in stock products
sns.barplot(data=df4,y='productName',x='quantityInStock')
plt.title('Barplot of High in Stock Products')
plt.ylabel('Product Name')
plt.xlabel('Quantity in stock');
```

```python
#top customers by revenue
sns.scatterplot(data=df5,y='contactLastName',x='revenue',hue='city',s=150)
plt.title('Top 10 Customers by Revenue and their city')
plt.ylabel('Customer Name')
plt.xlabel('Revenue');
```

```python
#bottom customers by revenue
sns.scatterplot(data=df6,y='contactLastName',x='revenue',hue='city',s=150)
plt.title('Bottom 10 customers by Revenue and their city')
plt.ylabel('Customer Name')
plt.xlabel('Revenue');
```

```python
#Top sales
sns.barplot(data=df7,y='productName',x='amount_sold',hue='productCode')
plt.title('Top sold products')
plt.ylabel('Product Name')
plt.xlabel('Amount sold');
```

```python
#Lowest sales
sns.barplot(data=df8,y='productname',x='amount_sold',hue='productCode')
plt.title('Least sold products')
plt.ylabel('Product Name')
plt.xlabel('Amount sold');
```

```python
#high credit limit customers
sns.barplot(data=df1,y='customerName',x='creditLimit')
plt.title('Barplot of Customer Credit Limits')
plt.ylabel('Customer Name')
plt.xlabel('Credit Limit');
```

```python
#low credit limit customers
sns.barplot(data=df2,y='customerName',x='creditLimit')
plt.title('Barplot of Customer Credit Limits')
plt.ylabel('Customer Name')
plt.xlabel('Credit Limit');
```
